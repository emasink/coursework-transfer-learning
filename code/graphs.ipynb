```python
from matplotlib import pyplot as plt
import numpy as np

def draw_graph(train_metric, val_metric, xlabel, ylabel):
    plt.figure(figsize=(10, 7))
    plt.plot(train_metric, label='Apmokymo duomenys')
    plt.plot(val_metric, label='Validacijos duomenys')
    plt.xticks(np.arange(0, 30, 1.0))
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.set_xlim(ymin=0)
    plt.legend()
    plt.show()
```

```python
lr = 0.001
import pickle
with open('SavedVariables/m2_metrics.pkl', 'rb') as f:
    metrics_by_epoch1 = pickle.load(f)

train_accuracies = [metrics_by_epoch1[i][0] for i in range(len(metrics_by_epoch1))]
train_losses = [metrics_by_epoch1[i][1] for i in range(len(metrics_by_epoch1))]
val_accuracies = [metrics_by_epoch1[i][2] for i in range(len(metrics_by_epoch1))]
val_losses = [metrics_by_epoch1[i][3] for i in range(len(metrics_by_epoch1))]
```

```python
draw_graph(train_accuracies, val_accuracies, 'Epochos', 'Tikslumas')
```

```python
draw_graph(train_losses, val_losses, 'Epochos', 'Paklaida')
```
